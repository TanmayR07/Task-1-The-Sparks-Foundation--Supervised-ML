### Plotting Sales Column

### Pie-Chart of Sub-Category Column

```python
plt.figure(figsize=(10,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

From the above pie-chart we can say that Paper and Blinders are the most occured sub-category 

## State wise Sales

```python
statewise=df.groupby(['State'])['Sales'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,10))
plt.title('State vs Sales')
statewise.plot.bar()
```

From above graph we can say that California has highest number of sales.

## State wise Profit

```python
statewise=df.groupby(['State'])['Profit'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, ax = plt.subplots(figsize=(20,10))
plt.title("State vs Profit")
statewise.plot.bar()
```

From the above graph we can say that California has the highest profit and then New York is in 2nd position. But the states like Oregon, Florida, Arizona, Tennessee, Colorado, Norht Carolina, Ilinois, Pennsylvania, Ohio and Texas have losses. Texas has incurred highest losses than any other state although Texas has made sales of approximately 175,000.

## Profit and Sales per Category

```python
s=df.groupby('Sub-Category')['Profit','Sales'].agg(['sum'])
s.plot.bar(rot=0, title="Total Profit and sales per category",figsize=(15,10));
plt.show(block=True);
```

Tabels have approximately 200,000 sales but is not a profit making good. Same goes for Supplies, as it is making sales of 49,000 yet it is making loss. Most profitable product for the store is Copier and then comes Phone and Accessories.

## Region wise Sub-Category

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region", data=df,palette="rainbow")
plt.show()
```

From the above graph we can say that people from west region buys more product.
